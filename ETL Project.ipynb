```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#1st Data Set imported
# load into a data frame
# zip code is showing as float with .0 decimals
facilities_file = "Wastewater_Treatment_Plants.csv"
facilities_info_df = pd.read_csv(facilities_file)
facilities_info_df.head()
```

```python
# convert the data frame

facilities_df = pd.DataFrame(facilities_info_df, dtype=object)
print(facilities_df)
```

```python
# new data frame shows zip code correctly
facilities_df.head()
```

```python
# changed column names
new_data = facilities_df.rename(columns = {"Plant Type": "plant_type", 
                                  "SPDES Permit Number":"SPDES_Permit", 
                                  "Facility Name": "Facility_name",
                                  "Ground or Surface":"Ground_Surface",
                                  "Average Design Hydraulic Flow":"Avg_DH_Flow",
                                  "Street":"Street_address",
                                  "Zip Code":"Zip_Code"}) 

new_data.head()
```

```python
 # Table #1, information about facilities
facilities_information = new_data[['SPDES_Permit','Zip_Code', 'Facility_name','plant_type','Ground_Surface','Avg_DH_Flow']].copy()
facilities_information.head()
```

```python
#save it into a csv file
facilities_information.to_csv('facilities_info.csv')
```

```python
# Table 2, locations
location_df = new_data[['Zip_Code','SPDES_Permit','Street_address','City','State','Country','Latitude','Longitude']].copy()
location_df.head()
```

```python
#save file for second table
location_df.to_csv('location_info.csv')
```

```python
# Load and Store the 2nd CSV into DataFrame
wastewater_file = "municipal-wastewater-treatment-plants.csv"
facility_treatment_df = pd.read_csv(wastewater_file)
facility_treatment_df.head()
```

```python
#3rd Table,water waste recycling details
waste_water_df = facility_treatment_df[['SPDES Number','County','Receiving Waterbody','Drainage Basin','Stream Class','Year Built','Year Updated','Plant Class','Population Served']].copy()
waste_water_df.head()
```

```python
#convert 2nd data frame
waste_fac_df = pd.DataFrame(waste_water_df, dtype=object)
print(waste_fac_df)
```

```python
new_data_2 = waste_fac_df.rename(columns = {"SPDES Number":"SPDES_Permit", 
                                  "Receiving Waterbody": "Receiving_Waterbody",
                                  "Drainage Basin":"Drainage_Basin",
                                  "Stream Class":"Stream_Class",
                                  "Year Built":"Year_Built",
                                  "Year Updated":"Year_Updated",
                                  "Plant Class":"Plant_Class",
                                  "Population Served":"Population_Served",
}) 

new_data_2.head()
```

```python
#save file for 3rd table
new_data_2.to_csv('wastewater_fac.csv')
```
